```python
#pair plot
#if there are many variables ,it is not convenient to draw scatter plots for each pairs of variables
#to understand the relationship . so a pair plot can be used to depict the relationships in a single diagram which can be plotted using pairplot() method .
#we will explore relationship of 4 variables SR-B,AVE,SOLD PRICE, SIXERS.
# the plot is drawn like a matrix and each row and column is respresented by that row and column variable .Each cell depicts the relationship between two variable represented by that row variable
```

```python
influential_feature=['SR_B','AVE','SIXERS','SOLD PRICE']
sea.pairplot(ipl_auction_df[influential_feature])
plt.show()
#for all the correlations,AVE and SIXERS seem to be highly correlated with SOLD PRICE   compare to strike rate batsman
```

```python
#correlation & heatmap
#correlation is used for measuring the strength and direction of the linear relationship two continuous random variables X and Y . It is a statistical measure that  indicate the extent to which two variables change together.
# a positive correlation: means the variables increase or decrease together 
# a negative correlation: means one variable increases, the other decreases 
#the correlation values lies between -1.0 and 1.0 . the signs indicates it is positive 
#-1 prefect negative correlation, 1 prefect positive correlation
# correlation values can be computed using corr()

influential_feature=['SR_B','AVE','SIXERS','SOLD PRICE']
ipl_auction_df[influential_feature].corr()
```

```python
sea.heatmap(ipl_auction_df[influential_feature].corr(),annot=True)
plt.title("Heatmap of correlation values")
plt.show()
#the color map scale is shown along the heatmap.Setting annot attribute equal to true print the correlation values in each box of the heatmap and improves the readability of the heatmap
#interpretation : the heatmap show that AVE and SIXER show positive correlation,while sr-b and SOLD price are not so strongly correlated .
#true
```

```python
sea.heatmap(ipl_auction_df[influential_feature].corr())
plt.title("Heatmap of correlation values")
plt.show()
```

```python
#pyplot
#a pie chart (or a circle chart) is a circular statistical graphics which is divided into slices to illustrate numerical proportion.
#the area of the chart is the total percentage of the given data.
#the area of slices of the pie represents the percentages of the parts of the data 
#with pyplot,we can use the pie() function to draw pie charts:
y=np.array([4,25,61,35])
plt.pie(y)
plt.show()
```

```python
#ques:Pie plot of average sold price percentage versus age 
SOLD_PRICE=[720250.000000,484534.883721,520178.571429]
my_labels='AGE_1','AGE_2','AGE_3'
plt.pie(SOLD_PRICE,labels=my_labels)
plt.title('Pie plot of average sold price percentage versus age')
plt.axis('equal')
plt.legend()
plt.show()
```
